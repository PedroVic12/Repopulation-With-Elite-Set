## Interface beta

```python
import easygui

def show_input_form(default_data):
    """
    Exibe um formulário de entrada com valores padrão e permite edição através de uma lista suspensa.
    
    Args:
        default_data (dict): Dicionário contendo os valores padrão para os campos.
    
    Returns:
        dict: Dicionário contendo os dados preenchidos pelo usuário.
    """
    data_dict = {}

    for field_name, default_value in default_data.items():
        # Verifica se o valor padrão não é uma lista, o que indica um campo de entrada
        if not isinstance(default_value, list):
            # Exibe um input para o campo
            value = easygui.enterbox(f"Digite o valor para '{field_name}':", field_name, default_value)
            data_dict[field_name] = value

    return data_dict

# Define os valores padrão para os campos
default_values = {
    "CROSSOVER": 0.9,
    "TAXA MUTACAO": 0.05,
    "GENERATIONS": 100,
    "POP_SIZE": 100,
    "IND_SIZE": 10,
    "evaluations": 0,
    "num_genes": 5
}

# Chama a função para exibir o formulário de entrada com valores padrão
input_data = show_input_form(default_values)

# Verifica se o formulário foi preenchido e exibe os dados
if input_data is not None:
    print("Dados preenchidos:")
    for key, value in input_data.items():
        print(f"{key}: {value}")
else:
    print("Operação cancelada pelo usuário.")

# Exibe a tela para escolha do tipo de otimização (minimização/máximação)
min_max_choice = easygui.buttonbox("Selecione o tipo de otimização:", "Minimize/Maximize", choices=["Minimize", "Maximize"])
print(f"Tipo de otimização selecionado: {min_max_choice}")

# Exibe a tela para escolha da função objetivo (Rastrigin ou Sphere)
function_choice = easygui.buttonbox("Selecione a função objetivo:", "Função Objetivo", choices=["Rastrigin", "Sphere"])
print(f"Função objetivo selecionada: {function_choice}")
```
